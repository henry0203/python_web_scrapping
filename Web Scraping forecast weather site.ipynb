## Forecast weather

```python
#https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.XtnbOJ4zYUo
import requests
from bs4 import BeautifulSoup
response = requests.get("https://forecast.weather.gov/MapClick.php?lat=37.7772&lon=-122.4168#.XtnbOJ4zYUo")
content = response.content
parser = BeautifulSoup(content, "html.parser")
seven_day = parser.find_all(id="seven-day-forecast-body")[0]  #or find() -> catch first element found
forecast_items = seven_day.find_all(class_="tombstone-container")
today = forecast_items[0]
print(today.prettify())
#seven_day_forecast = parser.select("#seven-day-forecast-body")[0].select(".tombstone-container")[1].text
#seven_day_forecast

#Get information from parsing object
period = today.find(class_="period-name").get_text()
short_desc = today.find(class_="short-desc").get_text()
temp = today.find(class_="temp").get_text()

print(period)
print(short_desc)
print(temp)

#Get information from img
img = today.find("img")
#print(img)
desc = img["title"] #img is a dictionnary, we get title with []
print(desc)

period_tags = seven_day.select(".tombstone-container .period-name")
#we go through all elements of period_tags and for each element applies get_text()
periods = [pt.get_text() for pt in period_tags]
print(periods)

short_descs = seven_day.select(".tombstone-container .short-desc")
#we go through all elements of period_tags and for each element applies get_text()
short_descs = [sd.get_text() for sd in short_descs]
print(short_descs)

temps = seven_day.select(".tombstone-container .temp")
#we go through all elements of period_tags and for each element applies get_text()
temps = [t.get_text() for t in temps]
print(temps)

descs = seven_day.select(".tombstone-container img")
#we go through all elements of period_tags and for each element applies get_text()
descs = [d["title"] for d in descs]
print(descs)
```

## Display results with Panda lib

```python
import pandas as pd
weather = pd.DataFrame({
    "period": periods,
    "short_desc": short_descs,
    "temp": temps,
    "desc": descs
})
weather
```
